# My 10 Propositions:

### Proposition 1 â€“ Ranking orders within each customer


-For each customer, rank their orders by TotalDue from highest to lowest. Show only orders that are in the top 5 per customer.

-USES ROW_NUMBER() + top N per group

````python
%%sql
USE AdventureWorks2019;

WITH RankedOrders AS (
    SELECT
        soh.CustomerID,
        soh.SalesOrderID,
        soh.OrderDate,
        soh.TotalDue,
        ROW_NUMBER() OVER (
            PARTITION BY soh.CustomerID
            ORDER BY soh.TotalDue DESC
        ) AS OrderRank
    FROM Sales.SalesOrderHeader AS soh
)
SELECT
    CustomerID,
    SalesOrderID,
    OrderDate,
    TotalDue,
    OrderRank
FROM RankedOrders
WHERE OrderRank <= 5
ORDER BY CustomerID, OrderRank;
````

### Proposition 2 â€“ Running total of sales by territory


-For each territory, list all orders ordered by date and compute the running total of TotalDue in that territory.

-USES Windowed SUM() with a frame for a running total

````python
%%sql
USE AdventureWorks2019;


SELECT
    soh.TerritoryID,
    soh.SalesOrderID,
    soh.OrderDate,
    soh.TotalDue,
    SUM(soh.TotalDue) OVER (
        PARTITION BY soh.TerritoryID
        ORDER BY soh.OrderDate, soh.SalesOrderID
        ROWS BETWEEN UNBOUNDED PRECEDING AND CURRENT ROW
    ) AS RunningTotal_Territory
FROM Sales.SalesOrderHeader AS soh
ORDER BY TerritoryID, OrderDate, SalesOrderID;

````

### Proposition 3: Orders as a percentage of customer total

-For each order, show what percentage of that customerâ€™s total spending this order represents.

-Uses Windowed SUM() with PARTITION BY to compute percent of group total.

````python
%%sql
USE AdventureWorks2019;

SELECT
    soh.CustomerID,
    soh.SalesOrderID,
    soh.OrderDate,
    soh.TotalDue,
    soh.TotalDue * 1.0 /
        SUM(soh.TotalDue) OVER (PARTITION BY soh.CustomerID) AS PercentOfCustomerTotal
FROM Sales.SalesOrderHeader AS soh
ORDER BY CustomerID, OrderDate;

````

### Proposition 4: Bucket customers into value tiers (NTILE)

-Compute each customerâ€™s lifetime total sales and divide customers into 4 value tiers (1 --> highest spenders, 4 --> lowest) using NTILE(4)

-NTILE() used for bucketing / segmentation.

````python
%%sql
USE AdventureWorks2019;

WITH CustomerTotals AS (
    SELECT
        soh.CustomerID,
        SUM(soh.TotalDue) AS LifetimeTotal
    FROM Sales.SalesOrderHeader AS soh
    GROUP BY soh.CustomerID
)
SELECT
    CustomerID,
    LifetimeTotal,
    NTILE(4) OVER (ORDER BY LifetimeTotal DESC) AS ValueTier
FROM CustomerTotals
ORDER BY LifetimeTotal DESC;
````

### Proposition 5: Monthly sales vs. 4-month moving average


-Aggregate total sales per calendar month and compute a 4-month moving average of monthly sales for the company.

-Moving average using a window frame (ROWS BETWEEN 2 PRECEDING AND CURRENT ROW).

````python
%%sql
USE AdventureWorks2019;

WITH MonthlySales AS (
    SELECT
        CONVERT(date, EOMONTH(soh.OrderDate)) AS EndOfMonth,
        SUM(soh.TotalDue) AS MonthlySales
    FROM Sales.SalesOrderHeader AS soh
    GROUP BY EOMONTH(soh.OrderDate)
)
SELECT
    EndOfMonth,
    MonthlySales,
    AVG(MonthlySales) OVER (
        ORDER BY EndOfMonth
        ROWS BETWEEN 3 PRECEDING AND CURRENT ROW
    ) AS MovingAvg_4Month
FROM MonthlySales
ORDER BY EndOfMonth;

````

### Proposition 6: Sales amount by year for each territory

-Show the total TotalDue for each territory, with one column per year (for example, 2011â€“2014). Territories should be rows, years should be columns.

-PIVOT : rotating row values into year columns.

````python
%%sql
Use AdventureWorks2019;

WITH TerritoryYearSales AS (
    SELECT
        soh.TerritoryID,
        YEAR(soh.OrderDate) AS OrderYear,
        SUM(soh.TotalDue)   AS YearlySales
    FROM Sales.SalesOrderHeader AS soh
    GROUP BY soh.TerritoryID, YEAR(soh.OrderDate)
)
SELECT
    TerritoryID,
    ISNULL([2011], 0) AS Sales2011,
    ISNULL([2012], 0) AS Sales2012,
    ISNULL([2013], 0) AS Sales2013,
    ISNULL([2014], 0) AS Sales2014
FROM TerritoryYearSales
PIVOT (
    SUM(YearlySales)
    FOR OrderYear IN ([2011], [2012], [2013], [2014])
) AS p
ORDER BY TerritoryID;
````

### Proposition 7:  Break down order cost components


-For each sales order, list SubTotal, TaxAmt, and Freight as separate rows with a column indicating the component name and another for the amount.

-USES UNPIVOT â€“ turning wide columns into long component rows.

````python
%%sql
USE AdventureWorks2019;

SELECT
    SalesOrderID,
    Component,
    Amount
FROM (
    SELECT
        SalesOrderID,
        SubTotal,
        TaxAmt,
        Freight
    FROM Sales.SalesOrderHeader
) AS src
UNPIVOT (
    Amount FOR Component IN (SubTotal, TaxAmt, Freight)
) AS u
ORDER BY SalesOrderID, Component;
````

### Proposition 8: GROUPING SETS: Territory, Year, and Grand Total


-Produce:
Sales by (TerritoryID, Year)
Sales by Year only
A single grand total for all territories and years

-GROUPING SETS â€“ multiple aggregation levels in one query.

````python
%%sql
USE AdventureWorks2019;

SELECT
    YEAR(soh.OrderDate) AS OrderYear,
    soh.TerritoryID,
    SUM(soh.TotalDue) AS TotalSales
FROM Sales.SalesOrderHeader AS soh
GROUP BY GROUPING SETS (
    (YEAR(soh.OrderDate), soh.TerritoryID),
    (YEAR(soh.OrderDate)),
    ()
)
ORDER BY OrderYear, TerritoryID;
````

### Proposition 9: ROLLUP: Year â†’ Month hierarchy with subtotals


-Show total sales by Year and Month, and use ROLLUP so that the result includes:

Monthly totals

Yearly totals

A final grand total

-ROLLUP â€“ automatic hierarchical subtotals (special case of GROUPING SETS).

````python
%%sql
USE AdventureWorks2019;

SELECT
    YEAR(soh.OrderDate)  AS OrderYear,
    MONTH(soh.OrderDate) AS OrderMonth,
    SUM(soh.TotalDue)    AS TotalSales
FROM Sales.SalesOrderHeader AS soh
GROUP BY ROLLUP (
    YEAR(soh.OrderDate),
    MONTH(soh.OrderDate)
)
ORDER BY OrderYear, OrderMonth;
````

### Proposition 10: Compare each product to the average price in its category



-For each product, show its list price and compare it to the average list price of all products in the same product subcategory.

-Windowed AVG() used for per-group comparison (product vs. category average).

````python
%%sql
USE AdventureWorks2019;

SELECT
    p.ProductID,
    p.Name AS ProductName,
    p.ProductSubcategoryID,
    p.ListPrice,
    AVG(p.ListPrice) OVER (
        PARTITION BY p.ProductSubcategoryID
    ) AS AvgPriceInSubcategory,
    p.ListPrice - AVG(p.ListPrice) OVER (
        PARTITION BY p.ProductSubcategoryID
    ) AS DifferenceFromAvg
FROM Production.Product AS p
WHERE p.ListPrice > 0
ORDER BY ProductSubcategoryID, ProductName;

````
